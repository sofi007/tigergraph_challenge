```python
list_ChemicalName = []

for chem in set_ChemicalName:
    nb_1 = chem.count('(')
    nb_2 = chem.count(')')
    if("+" in chem):
        chem = chem.replace("+","Â§")
    if("(3Z)-3-(1H-PYRROL-2-YL)-METHYLLIDENE)-1-(1-(PIPERIDYLMETHYL)-1,3-DIHYDRO-2H-INDOL-2-ONE MESYLATE" in chem):
        chem = "((3Z)-3-(1H-PYRROL-2-YL)-METHYLLIDENE)-1-(1-(PIPERIDYLMETHYL)-1,3-DIHYDRO-2H-INDOL-2-ONE MESYLATE)"
        
    if (nb_1 != nb_2):
        
        diff = nb_1 - nb_2
        if (diff > 0):
            for i in range(0,diff):
                chem = chem + ")"
            
        diff = nb_2 - nb_1
        if (diff > 0):
            for i in range(0,diff):
                chem = "(" + chem
            
    chem = chem.replace("(","<")    
    chem = chem.replace(")",">")    
    list_ChemicalName.append(chem)
        


print(len(set_ChemicalName))
print(len(list_ChemicalName))
```

```python
def write_node(folder, filename, set_nodes_prop):
    
    file_write = open(folder+filename, 'w')
    sep=","
    for t in set_nodes_prop:
        string = t[0] +sep+ t[1]+"\n"
        file_write.write(string)
    file_write.close()
```

```python
folder = ""
filename = "set_gene.csv"

write_node(folder, filename, set_gene_id)
```
